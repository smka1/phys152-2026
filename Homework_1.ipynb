```python
from torch import nn, optim

# I choose 2 hidden layers, and 8 then 3 hidden variables. Also because of the nature I chose Tanh as my activation
from torch import nn, optim
complex_model = nn.Sequential(
    nn.Linear(1, 8),
    nn.Tanh(),
    nn.Linear(8, 3),
    nn.Tanh(),
    nn.Linear(3, 1)
)
lr =0.1
epochs = 100000

# TO-DO: update the learning rate
criterion = nn.MSELoss()
optimizer = optim.SGD(complex_model.parameters(), lr=lr)
```

```python
# TO-DO: update number of epochs
for epoch in range(epochs):
    # Forward pass: Compute predicted y by passing x to the model
    pred_y = complex_model(x_data_tensor)
    loss = criterion(pred_y, y_data_tensor)

    # Zero gradients, perform a backward pass,
    # and update the weights.
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10000 == 0:
        print(f"epoch {epoch}, loss {loss.item()}")
```
